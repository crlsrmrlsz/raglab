```python
import weaviate
```

## test connection

```python

print("Connecting to Weaviate...")
client = weaviate.connect_to_local()

try:
    print(f"✓ Connected!")
    print(f"✓ Ready: {client.is_ready()}")
    
    meta = client.get_meta()
    print(f"✓ Version: {meta.get('version')}")
    
    collections = client.collections.list_all()
    print(f"✓ Collections: {list(collections.keys())}")
    
    print("\n🎉 Weaviate is working correctly!")
    
finally:
    client.close()
    print("Connection closed.")
```

## see collections

```python
print("Connecting to Weaviate...")
client = weaviate.connect_to_local()
```

```python
#List all collections:
client.collections.list_all().keys()
```

```python
from pprint import pprint

#Get collection details (schema):
collection = client.collections.get("RAG_section_embed3large_v1")
config = collection.config.get()
pprint(config.to_dict())
```

## number of objects

```python
collection = client.collections.get("RAG_section_embed3large_v1")
number = collection.aggregate.over_all(total_count=True)
number
```

## view sample objects

```python
#View sample objects:
collection = client.collections.get("RAG_section800_embed3large_v1")
response = collection.query.fetch_objects(limit=2, include_vector=False)
for item in response.objects:
  pprint(item.properties)
  print("---")
```

## delete collection

```python
#Delete a collection:
client.collections.delete("RAG_section800_embed3large_v1")
```

## delete all collections

```python
client = weaviate.connect_to_local()

# List existing collections
collections = list(client.collections.list_all().keys())
print(f"Found {len(collections)} collections: {collections}")

# Delete all
for name in collections:
  client.collections.delete(name)
  print(f"Deleted: {name}")

client.close()
print("Done")
```
